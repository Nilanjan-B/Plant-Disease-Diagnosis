```python
import os
import matplotlib.pyplot as plt
import tensorflow as tf

# Define the path to your raw data
data_dir = 'data/raw' # Ensure this matches where you extracted PlantVillage

# Standardizing the input for our future model
BATCH_SIZE = 32
IMAGE_SIZE = (256, 256) # Resizing all leaves to 256x256 pixels

# Load the dataset using TensorFlow
dataset = tf.keras.utils.image_dataset_from_directory(
    data_dir,
    shuffle=True,
    image_size=IMAGE_SIZE,
    batch_size=BATCH_SIZE
)

# Extract the class names (the specific plant diseases)
class_names = dataset.class_names
print(f"Found {len(class_names)} distinct plant/disease categories.")

# Visualize a batch of leaves
plt.figure(figsize=(12, 12))
for images, labels in dataset.take(1): # Take 1 batch of 32 images
    for i in range(9): # Show the first 9
        ax = plt.subplot(3, 3, i + 1)
        plt.imshow(images[i].numpy().astype("uint8"))
        plt.title(class_names[labels[i]])
        plt.axis("off")
plt.show()
```
